```python
import os
from langchain_core.prompts import PromptTemplate
from langchain_huggingface.llms import HuggingFacePipeline
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, pipeline

# Setup
hf_token = "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"
os.environ["HUGGINGFACEHUB_API_TOKEN"] = hf_token
```

```python
# Pipeline Steps
# Step 1: Takes user input and reformulates it into a clear query
# Step 2: Generates a comprehensive detailed response
# Step 3: Creates a concise summary of the detailed response
model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

text_gen = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=512,
    temperature=0.7
)
llm = HuggingFacePipeline(pipeline=text_gen)
```

```python
# User input
user_input = "where is paris"

print("ORIGINAL INPUT:", user_input)
```

```python
# STEP 1: Reformulate
print("STEP 1: REFORMULATING QUERY")
print("-"*60)
reformulate_prompt = f"Rewrite this as a clear formal question: {user_input}"
reformulated = llm.invoke(reformulate_prompt)
print(reformulated + "\n")
```

```python
# STEP 2: CREATING SUMMARY
print("STEP 2: CREATING SUMMARY")
print("-"*60)
summary_prompt = f"Explain in detail: {reformulated}"
summary = llm.invoke(summary_prompt)
print(summary + "\n")


print("PIPELINE COMPLETE!")
print("-"*60)
```

```python
# Final output
print("-"*60)
print("SUMMARY OF PIPELINE:")
print("-"*60)
print(f"Original: {user_input}")
print(f"Reformulated: {reformulated}")
print(f"Summary: {summary}")
```
